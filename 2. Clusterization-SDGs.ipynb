```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('max_columns', None)
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PowerTransformer
from sklearn.preprocessing import Normalizer
from sklearn import preprocessing
```

```python
df2 = df
df3 = df[['title','author','year','keywords']]
df2 = df2.drop(['title','DOI (se houver)','author','year','abstract','word_count','keywords'],axis=1)
```

```python
ssd = []
K = range(1,30)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(df2)
    ssd.append(km.inertia_)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.plot(K, ssd, 'bx-')
plt.xlabel('k')
plt.ylabel('ssd')
plt.title('Elbow Method For Optimal k')
#plt.show()
```

```python
kmeans = KMeans(n_clusters=4, n_init=25, random_state=0)
model = kmeans.fit(df2)
```

```python
pred = model.labels_
df2['cluster'] = pred
df2.head()
```

```python
df_total = pd.concat([df2, df3], axis=1)
df_total.head()
```

```python
df_0 = df_total[df_total['cluster'] == 0]
df_1 = df_total[df_total['cluster'] == 1]
df_2 = df_total[df_total['cluster'] == 2]
df_3 = df_total[df_total['cluster'] == 3]
```
